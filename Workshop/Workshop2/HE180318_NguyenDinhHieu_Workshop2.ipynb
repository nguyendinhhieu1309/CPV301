```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
from IPython.display import display, clear_output
```

## Function1


```python
def load_image(file_path):
    return cv2.imread(file_path)

def balance_colors(image, red_intensity, green_intensity, blue_intensity):
    balanced_image = np.copy(image)
    balanced_image[:, :, 0] = np.clip(image[:, :, 0] * (blue_intensity / 255.0), 0, 255)
    balanced_image[:, :, 1] = np.clip(image[:, :, 1] * (green_intensity / 255.0), 0, 255)
    balanced_image[:, :, 2] = np.clip(image[:, :, 2] * (red_intensity / 255.0), 0, 255)
    return balanced_image

def update_color_balance(file_path, red_intensity, green_intensity, blue_intensity):
    image = load_image(file_path)
    balanced_image = balance_colors(image, red_intensity, green_intensity, blue_intensity)
    
    plt.figure(figsize=(8, 8))
    plt.imshow(cv2.cvtColor(balanced_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

default_image_path = "hola.jpg"

red_slider = widgets.IntSlider(value=128, min=0, max=255, description='Red')
green_slider = widgets.IntSlider(value=128, min=0, max=255, description='Green')
blue_slider = widgets.IntSlider(value=128, min=0, max=255, description='Blue')

interact(update_color_balance, file_path=widgets.fixed(default_image_path), red_intensity=red_slider, green_intensity=green_slider, blue_intensity=blue_slider)
```

## Function2

```python
image = cv2.imread("TUM.jpg", cv2.IMREAD_GRAYSCALE)
def calculate_and_display_histogram(image):
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])
    return hist

def equalize_histogram(image):
    equalized_image = cv2.equalizeHist(image)
    return equalized_image

def update_histogram_equalization():
    equalized_image = equalize_histogram(image)
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    plt.subplot(1, 2, 2)
    plt.imshow(equalized_image, cmap='gray')
    plt.title('Equalized Image')
    plt.axis('off')   
    plt.tight_layout()
    plt.show()
    histogram = calculate_and_display_histogram(equalized_image)
    plt.figure(figsize=(8, 4))
    plt.plot(histogram)
    plt.title('Histogram')
    plt.show()

update_histogram_equalization()
```

## Function3

```python
image = cv2.imread("TUM.jpg", cv2.IMREAD_GRAYSCALE)
def median_filter(image, kernel_size):
    filtered_image = cv2.medianBlur(image, kernel_size)
    return filtered_image

kernel_size = 3
filtered_image = median_filter(image, kernel_size)

plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(filtered_image, cmap='gray')
plt.title('Median Filtered Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

## Function4

```python
image = cv2.imread("hola.jpg", cv2.IMREAD_GRAYSCALE)
def mean_filter(image, kernel_size):
    filtered_image = cv2.blur(image, (kernel_size, kernel_size))
    return filtered_image
kernel_size = 3  
filtered_image = mean_filter(image, kernel_size)

plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(filtered_image, cmap='gray')
plt.title('Mean Filtered Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

## Function5

```python
image = cv2.imread("camera.jpg", cv2.IMREAD_COLOR)
def gaussian_smoothing(image, kernel_size, sigma):
    smoothed_image = cv2.GaussianBlur(image, (kernel_size, kernel_size), sigma)
    return smoothed_image
kernel_size = 5  
sigma = 0  
smoothed_image = gaussian_smoothing(image, kernel_size, sigma)

plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(smoothed_image, cv2.COLOR_BGR2RGB))
plt.title('Smoothed Image (Gaussian Smoothing)')
plt.axis('off')
plt.tight_layout()
plt.show()
```
